# Condicionales

Utilizan condiciones para ejecutar (o no ejecutar) código.

## if

Se usa para determinar si ejecutar un bloque de código o no.

Formato:
```
if (condición):
    (código)
```

````python
a = 1
b = 2

if a < b:
    print("Hola")
````

## else

Se utiliza después de un if, y en caso de que la condición sea falsa se ejecuta.

Formato:
```
if (condición):
    (código)
else:
    (código)
```

````python
a = 1
b = 2

if a > b:
    print("Hola 1")
else:
    print("Hola 2")
````

## elif

Se utiliza después de un if, y en caso de que la condición del if sea falsa, chequea su condición, y si es verdadera se ejecuta, en caso de no serlo pasa al siguiente.

Formato:
```
if (condición):
    (código)
elif (condición):
    (código)
```

````python
a = 1
b = 2

if a > b:
    print("Hola 1")
elif a < b:
    print("Hola 2")
else:
    print("Hola 3")
````
