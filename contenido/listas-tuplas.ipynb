# Listas

Las listas son un conjunto ordenado de objetos, empezando desde 0.

Formato: `[a, b, c, ...]`

## Operatoria

### [a]

Sirve para obtener el elemento número a de la lista.

```python
# Pos    0  1  2  3
lista = [1, 2, 3, 4]

lista[3]
```

### [a:b:c]

Sirve para modificar la lista, cambiando según los parámetros:
- a: cambia donde empieza la lista. Inclusivo
- b: cambia donde termina la lista. Exclusivo
- c: omite cada c elementos en la lista. SE APLICA PRIMERO.

Cada una de estas se puede dejar en blanco, siempre que se mantengan los ":".

```python
# Pos    0  1  2   3   4   5
lista = [7, 5, 13, 24, 20, 18]

print(lista[3:])    # 1: [24, 20, 18]
print(lista[::-3])  # 2: [18, 13]
print(lista[:3:2])  # 3: [24] [7, 13]
print(lista[2::-2]) # 4: [13, 7]
print(lista[::-1])  # 5: [18, 20, 24, 13, 5, 7]

lista.sort()

print(lista[0], lista[len(lista) - 1], sep="\n")
```

## Funciones de lista

### append(x)

Añade un elemento x al último lugar de la lista.

```python
lista = [2, 1, 4]
lista.append(9)

lista
```

### remove(x)

Quita la primera ocurrencia de un elemento en la lista.

```python
lista = [2, 5, 3, 2, 7]
lista.remove(2)

lista
```

### insert(x, y)

Agrega el elemento y, tomando la posición x.

```python
lista = [2, 5, 3, 2, 7]
lista.insert(3, 9)

lista
```

### index(x)

Da la posición de la primera ocurrencia del elemento x en la lista.

```python
lista = [2, 5, 3, 9, 2, 7]

lista.index(2)
```

### pop(x)

Por defecto, elimina el último elemento de la lista. Con un argumento, elimina el elemento x de la lista.

Devuelve el valor que quitó.

```python
lista = [2, 5, 3, 9, 2, 7]

lista.pop(4)
```

### count(x)

Cuenta las ocurrencias de x en la lista.

```python
lista = [2, 5, 3, 5, 9, 2, 7, 5]

lista.count(5)
```

### sort()

Ordena la lista, dependiendo del tipo de dato.
- str o chr: orden alfabético.
- int o float: orden en los reales.

```python
lista = [2, 5, 3, 5, 9, 2, 7, 5]
lista.sort()

lista
```

### reverse()

Da vuelta la lista.

```python
lista = [2, 5, 3, 5, 9, 2, 7, 5]
lista.reverse()

lista
```

### clear()

Quita todos los elementos.

```python
lista = [2, 5, 3, 5, 9, 2, 7, 5]
lista.clear()

lista
```
