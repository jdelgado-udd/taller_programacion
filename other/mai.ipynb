```python
dict_criterios = {"largo": 0,"inicio": 0,"digito": 0,"caracteres": 0,"prohibidos": 0}
acumulador = 0
validos = 0
salir = False

while not salir:
    usuario = input("ingrese el usuario: ")

    if "salir" == usuario.lower():
        salir = True
        continue

    if len(usuario) == 0:
        continue


    acumulador += 1
    lenusuario = len(usuario)
    numeros = 0
    invalido = 0
    for i in range(lenusuario):

        letra = usuario[i]
        if letra.isdigit():
            numeros += 1

        if not letra.isalnum() and (letra not in "-._"):
            invalido += 1

    if lenusuario >= 6 and lenusuario <= 20 and numeros >=1 and invalido == 0 and usuario[0].isalpha() and "admin" not in usuario.lower() and "root" not in usuario.lower():
        validos += 1

    if lenusuario < 6 or lenusuario > 20:
        dict_criterios["largo"] += 1

    if not usuario[0].isalpha():
        dict_criterios["inicio"] += 1

    if numeros < 1:
        dict_criterios["digito"] += 1

    if invalido > 0:
        dict_criterios["caracteres"] += 1

    if "admin" in usuario.lower() or "root" in usuario.lower():
        dict_criterios["prohibidos"] += 1

if acumulador > 0:
    print(f"usuarios validos: {validos} de {acumulador}")
    print(f"usuarios invalidos: {acumulador - validos}")
    print(f"usuarios invalidos por criterios:")
    print(f"largo: {dict_criterios['largo']}")
    print(f"inicio: {dict_criterios['inicio']}")
    print(f"digito: {dict_criterios['digito']}")
    print(f"caracteres: {dict_criterios['caracteres']}")
    print(f"prohibidos: {dict_criterios['prohibidos']}")
else:
    print("no hay")


```
