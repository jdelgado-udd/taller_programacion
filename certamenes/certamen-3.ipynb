## CÓDIGO

```python
class Bodega:
    total_bodegas = 0
    
    def __init__(self, capacidad, tarifa_base):
        if capacidad <= 0 or tarifa_base <= 0:
            raise ValueError("todos los valores deben ser mayores a 0")
        
        self.capacidad = capacidad
        self.tarifa_base = tarifa_base
        self._stock = {}
        Bodega.total_bodegas += 1

    def guardar(self, producto, cantidad):
        if cantidad <= 0:
            raise ValueError

        actual = cantidad

        if producto in self._stock:
            actual += self._stock[producto]

        if actual > self.capacidad:
            raise RuntimeError("capacidad excedida")

        self._stock[producto] = actual

    def retirar(self, producto, cantidad):
        if producto not in self._stock:
            raise ValueError(f"'{producto}' no está en la bodega")

        if cantidad <= 0:
            raise ValueError("la cantidad debe ser positiva")

        actual = self._stock[producto] - cantidad

        if actual < 0:
            raise ValueError("stock insuficiente")

        if actual == 0:
            del self._stock[producto]
        else:
            self._stock[producto] = actual

    def total_unidades(self):
        return sum(self._stock.values())

    def espacio_libre(self):
        return self.capacidad - self.total_unidades()

    def productos(self):
        return list(self._stock.keys())

    def costo_almacenaje(self):
        return self.total_unidades() * self.tarifa_base

    def __repr__(self):
        return f"Bodega(${self.costo_almacenaje()})"

    def __str__(self):
        return f"Bodega(${self.costo_almacenaje()})"

class BodegaRefrigerada(Bodega):
    def __init__(self, capacidad, tarifa_base, recargo):
        super().__init__(capacidad, tarifa_base)
        
        if recargo <= 0:
            raise ValueError("recargo debe ser mayor a 0")

        self.recargo = recargo
        # NO INCREMENTAR BODEGAS PORQUE INIT HEREDADO YA LO HACE

    def costo_almacenaje(self):
        return self.total_unidades() * (self.tarifa_base + self.recargo)

    def __repr__(self):
        return f"BodegaRefrigerada(${self.costo_almacenaje()})"

    def __str__(self):
            return f"BodegaRefrigerada(${self.costo_almacenaje()})"

class BodegaPeligrosa(Bodega):
    def __init__(self, capacidad, tarifa_base, cargo_fijo):
        super().__init__(capacidad, tarifa_base)
        
        if cargo_fijo <= 0:
            raise ValueError("cargo_fijo debe ser mayor a 0")

        self.cargo_fijo = cargo_fijo
        # NO INCREMENTAR BODEGAS PORQUE INIT HEREDADO YA LO HACE

    def costo_almacenaje(self):
        return self.cargo_fijo * len(self.productos()) + self.total_unidades() * self.tarifa_base

    def __repr__(self):
        return f"BodegaPeligrosa(${self.costo_almacenaje()})"

    def __str__(self):
        return f"BodegaPeligrosa(${self.costo_almacenaje()})"
```
