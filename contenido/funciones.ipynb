# Funciones

Las funciones son pedazos de código reutilizables, se utilizan para poder usar continuamente un proceso que no queremos escribir miles de veces.

Lo aplicas cuando ocupas un proceso específico varias veces, y buscas acortar las líneas de código.

```python
import random

x = 10
a = 1
b = 50

lista = []

for i in range(x):
    c = random.randint(a, b)
    lista.append(c)

lista
```

```python
def generar_lista_aleatoria(cantidad_numeros, intervalo_a, intervalo_b):
    lista = []

    for i in range(cantidad_numeros):
        lista.append(random.randint(intervalo_a, intervalo_b))

    return lista

generar_lista_aleatoria(x, a, b)
```

```python
def funcion(argumento1, argumento2):
    print(argumento1, argumento2)
    return f"Hola mai {argumento2 - argumento1}"

llamada = funcion(1, 4)
print(f"Llamada: {llamada}")
```

```python
lista_temp_celsius = [22, 19, 16, 13, 10, 7, 4, 1]
lista_temp_farenheit = []

def a_farenheit(c):
    return (c * 9/5) + 32

for temp in lista_temp_celsius:
    lista_temp_farenheit.append(a_farenheit(temp))

print(lista_temp_celsius, lista_temp_farenheit)
```

```python
temperatura = float(input("Ingrese la temperatura: ")) # 20
unidad = input("Ingrese la unidad (C, K o F): ") # K
conversion = input("Ingrese la unidad a convertir (C, K o F): ") # F
resultado = 0

if unidad != conversion:
    normalizado = temperatura # normalizado = 20

    if unidad == "K":
        normalizado = normalizado - 273.15 # normalizado = 20 - 273.15 = -173.15
    elif unidad == "F":
        normalizado = (normalizado - 32) * 5/9

    if conversion == "K":
        resultado = normalizado + 273.15
    elif conversion == "F":
        resultado = (normalizado * 9/5) + 32 # resultado = (-173.15 * 9/5) + 32 = -423.669
    else:
        resultado = normalizado
else:
    resultado = temperatura

print(f"El resultado es {conversion}° {resultado}.")
```

```python
tabla_desde_c = {
    "C": lambda x: x,
    "K": lambda x: x + 273.15,
    "F": lambda x: (x * 9/5) + 32
}

tabla_a_c = {
    "C": lambda x: x,
    "K": lambda x: x - 273.15,
    "F": lambda x: (x - 32) * 5/9
}

temperatura = float(input("Ingrese la temperatura: ")) # 20
unidad = input("Ingrese la unidad (C, K o F): ") # K
conversion = input("Ingrese la unidad a convertir (C, K o F): ") # F
resultado = tabla_desde_c[conversion](tabla_a_c[unidad](temperatura))

print(resultado)
```

```python
def digito_romano(n, tabla): # n = 9, tabla = ["X", "L", "C"] | n = 5, tabla = ["I", "V", "X"]
    if n == 0:
        return ""

    if n == 4:
        return tabla[0] + tabla[1]

    if n == 9: # Sí
        return tabla[0] + tabla[2] # XC

    if n == 10:
        return tabla[2]

    if n >= 5: # Sí
        return tabla[1] + tabla[0] * (n - 5) # V

    if n >= 1:
        return tabla[0] * n

    return None

def numero_romano(n): # n = 95
    centenas = n // 100
    decenas = (n - n // 100 * 100) // 10 # decenas = 9
    unidades = (n - n // 10 * 10) // 1 # unidades = 5

    return digito_romano(centenas, ["C", "D", "M"]) + digito_romano(decenas, ["X", "L", "C"]) + digito_romano(unidades, ["I", "V", "X"])

numero_romano(795)
```
