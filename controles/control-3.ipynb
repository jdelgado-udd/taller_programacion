## Pregunta 1

Programa para calcular promedios:
- 3 notas parciales V
- Formato: nombre nota1 nota2 nota3 V
- Salida con "fin" V
- Nombre almacenado con primera en mayúscula V
- Nota entre 1 y 7 V
- Descarte de formato invalido con mensaje V
- Crear diccionario nombre - promedio V
- Reporte con cantidad de estudiantes, mejor promedio, cantidad aprobados >= 4, promedio general (2 decimales) V

```python
print("""

Calculadora de Notas

Formato de registro: <nombre> <nota1> <nota2> <nota3>
Salida:              fin
""")

promedios_estudiantes = {}
cantidad_aprobados = 0
mayor_promedio = None

while True:
    entrada = input("Ingrese un registro de estudiante: ")
    datos = entrada.split(" ")

    if entrada.lower() == "fin":
        break

    if len(datos) != 4:
        print("Por favor siga el formato indicado.")
        continue

    nombre = datos[0].capitalize()
    nota1 = datos[1]
    nota2 = datos[2]
    nota3 = datos[3]

    try:
        nota1 = float(nota1)
        nota2 = float(nota2)
        nota3 = float(nota3)
    except:
        print("Las notas deben de ser números decimales.")
        continue

    if 7 < nota1 or nota1 < 1 or 7 < nota2 or nota2 < 1 or 7 < nota3 or nota3 < 1:
        print("El rango de notas debe estar entre 1 y 7.")
        continue

    promedio = (nota1 + nota2 + nota3) / 3

    if mayor_promedio is None or promedio > promedios_estudiantes[mayor_promedio]:
        mayor_promedio = nombre

    if promedio >= 4:
        cantidad_aprobados += 1
    
    promedios_estudiantes[nombre] = promedio

# Reporte con cantidad de estudiantes, mejor promedio, cantidad aprobados >= 4, promedio general (2 decimales)

if len(promedios_estudiantes) < 1:
    print("No se ha generado un reporte debido a que no hay estudiantes.")
else:
    promedios_lista = list(promedios_estudiantes.values())
    
    print(
        "",
        "REPORTE DE NOTAS",
        f" Cantidad de estudiantes: {len(promedios_estudiantes)}",
        f" Mejor promedio: {mayor_promedio} ({promedios_estudiantes[mayor_promedio]})",
        f" Cantidad de aprobados: {cantidad_aprobados}",
        f" Promedio general: {round(sum(promedios_lista) / len(promedios_lista), 2)}",
        sep="\n"
    )
```

## Pregunta 2

Programa validador de contraseñas:
- Leer contraseñas de usuario una a una V
- Salida con "salir" independiente mayúsculas V
- Vacío = no evaluar V
- Al menos 10 caracteres V
- 2 mayúsculas V
- 2 dígitos V
- 1 caracter especial "!@#$%);" V
- no contener "password" ni "1234" independiente upper V
- Solo válida con 5 criterios V
- Conteo de válidas e invalidas V
- Diccionario de criterios fallidos para invalidas nombres largo, mayusculas, digitos, especial, prohibidos V
- Reporte cantidad de contraseñas validas, invalidas y fallas por criterio

```python
print("""

Validador de Contraseñas

 (Para salir del programa, ingrese "salir")
""")

contador_validas = 0
contador_invalidas = 0

criterios = { "largo": 0, "mayusculas": 0, "digitos": 0, "especial": 0, "prohibidos": 0 }

while True:
    entrada = input("Ingrese una contraseña: ")

    if len(entrada) < 1:
        print("Por favor, ingrese una contraseña.")
        continue

    if entrada.lower() == "salir":
        break

    largo = len(entrada)
    mayusculas = 0
    digitos = 0
    especiales = 0
    prohibidos = "password" in entrada.lower() or "1234" in entrada

    for character in entrada: # True numéricamente es tratado como 1, mientras que False es tratado como 0.
        mayusculas += character.isupper()
        digitos += character.isdigit()
        especiales += character in "!@#$%);"

    valida = True

    if largo < 10:
        criterios["largo"] += 1
        valida = False

    if mayusculas < 2:
        criterios["mayusculas"] += 1
        valida = False

    if digitos < 2:
        criterios["digitos"] += 1
        valida = False

    if especiales < 1:
        criterios["especiales"] += 1
        valida = False

    if prohibidos:
        criterios["prohibidos"] += 1
        valida = False

    if valida:
        contador_validas += 1
    else:
        contador_invalidas += 1

contador_totales = contador_validas + contador_invalidas

if contador_totales == 0:
    print("No se ha generado un reporte puesto que no se han ingresado contraseñas.")
else:
    print(
        "",
        "REPORTE DE SEGURIDAD DE CONTRASEÑAS",
        f" Válidas: {contador_validas} / {contador_totales}",
        f" Inválidas: {contador_invalidas} / {contador_totales}",
        f" Criterios:",
        sep="\n"
    )

    for criterio in criterios:
        print(f" - {criterio}: {criterios[criterio]} / {contador_invalidas}")
```
