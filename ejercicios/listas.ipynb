## Ejercicio 5

Una clínica atiende 12 horas al día. Pide n y luego n entradas con hora_llegada (0–23) de
pacientes.
* Cuenta cuántos llegaron fuera del horario (antes de 8 o después de 19).
* Calcula el porcentaje fuera de horario sobre n (2 decimales).
* Reporta la hora con más llegadas (modo de 0–23, si empata, la menor).

Salida:
* Fuera de horario: 3 (30.00%)
* Hora pico: 10
* Restricciones: No usar collections.Counter.

```python
cantidad_pacientes = int(input("Ingrese la cantidad de pacientes: "))
indice = 0

horas = []
fuera_horario = 0

while indice < cantidad_pacientes:
    hora_llegada = int(input(f"Ingrese la hora de llegada del paciente #{indice + 1}: "))

    if hora_llegada < 8 or hora_llegada > 19:
        fuera_horario += 1

    horas.append(hora_llegada)

    indice += 1

horas.sort()

indice = 1

inicio_mayor = 0
repeticion_mayor = 0
inicio = 0
repeticion = 0

while indice < cantidad_pacientes:
    if horas[indice - 1] == horas[indice]:
        repeticion += 1
    else:
        repeticion = 0
        inicio = indice

    if repeticion > repeticion_mayor:
        inicio_mayor = inicio

    indice += 1

hora_pico = horas[inicio_mayor]

print(
    f"Fuera de horario: {fuera_horario}",
    f"Fuera de horario %: {fuera_horario / cantidad_pacientes}",
    f"Hora pico: {hora_pico}",
    sep="\n"
)
```

Pide 5 números al usuario, guárdalos en una lista y luego muestra:
la lista completa,
el número mayor,
el número menor,
el promedio.

Puedes usar funciones como max, min y sum, o hacerlo manualmente.
Habilidad que practica:
Almacenamiento de datos y recorrido de listas.

```python
lista = []
prom = 0

while len(lista) < 5:
    entrada = input(f"Ingrese el número #{len(lista)}: ")

    try:
        entrada = float(entrada)
    except:
        print("Entrada invalida, por favor ingrese un número válido.")
        continue

    lista.append(entrada)
    prom = prom + entrada

lista.sort()
prom = prom / 5

print(f"Menor: {lista[0]}", f"Mayor: {lista[4]}", f"Promedio: {prom}", sep="\n")
```

cuántas palabras tiene,

la última palabra,
cuántas veces aparece una palabra específica que el usuario quiera buscar.

Usa split().

Manejo de strings y listas.

```python
frase = input("Ingrese una frase: ")

if frase.endswith("."):
    frase = frase[:-1]

palabras = frase.split(" ")

mensaje = ""

print(
    f"Palabras: {len(palabras)}",
    f"Primera palabra: {palabras[0]}",
    f"Última palabra: {palabras[len(palabras) - 1]}",
    sep="\n"
)

if input("Desea buscar una palabra? (V / F)").lower() == "v":
    palabra = input("Ingrese la palabra: ").lower()
    print(f"Ocurrencias: {frase.lower().split(" ").count(palabra)}")
```
