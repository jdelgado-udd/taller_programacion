# Ejercicios de diccionarios

## Ejercicio 13

Un edificio tiene 3 departamentos.

Datos: numero de depto, consumo (kWh) y tipo de cliente (residencial o comercial)

Reglas: Valor kWh
- Residencial = 120
- Comercial = 180

Mostrar gastos:
- total del edificio en pesos
- depto con mayor consumo
- promedio de consumo de residenciales
- si consumo total es mayor 5000 kWh, mostrar "sobrecarga"

Validación: consumo mayor igual 0


```python
clientes = {}
indice = 0

suma_gasto_total = 0
depto_mayor_consumo = 0
suma_consumo_residenciales = 0
cantidad_residenciales = 0
suma_consumo = 0

while indice < 3:
    depto_num = input("Por favor, ingrese el número del depto: ")
    depto_consumo = int(input("Ingrese su consumo en kWh: "))
    depto_comercial = input("Es su depto comercial? ").lower().startswith("s")
    depto_valor = 0

    if depto_comercial:
        depto_valor = 180
    else:
        depto_valor = 120

    clientes[depto_num] = (depto_consumo, depto_comercial, depto_consumo * depto_valor)

    suma_gasto_total = suma_gasto_total + (depto_valor * depto_consumo)
    suma_consumo = suma_consumo + depto_consumo

    if not depto_comercial:
        suma_consumo_residenciales = suma_consumo_residenciales + depto_consumo
        cantidad_residenciales = cantidad_residenciales + 1

    if depto_mayor_consumo not in clientes or clientes[depto_mayor_consumo][0] < depto_consumo:
        depto_mayor_consumo = depto_num

    indice = indice + 1

mensaje = ""

if suma_consumo > 5000:
    mensaje = "ALERTA: Sobrecarga"

print(
    f"Total: ${suma_gasto_total}",
    f"Mayor consumo: Depto {depto_mayor_consumo} ({clientes[depto_mayor_consumo]} kWh)",
    f"Promedio consumo (Residencial): {suma_consumo_residenciales / cantidad_residenciales}",
    mensaje,
    sep="\n"
)
```
