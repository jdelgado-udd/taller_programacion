# Ciclos

Son bloques de código que se repiten.

## while

El ciclo while utiliza una condición, y mientras esta sea verdadera ejecuta el código.

Formato:
```
while (condición):
    (código)
```

````python
a = 1
b = 8

while a + 1 != b:
    a += 1

print(a)
````

## for

Recorre el contenido de una lista, tupla.

Formato:
```
for (nombre de variable) in (lista / tupla):
    (código)
```

````python
lista = ["Joaco", "Mai", "Javi", "Lopez", "Emi"]

for nombre in lista:
    print(nombre)
````

El for también se utiliza mucho con la función "range(int, int)", que te entrega una lista con números en el rango especificado tal que [a, b[.

````python
for indice in range(2, 7):
    print(indice)
````
