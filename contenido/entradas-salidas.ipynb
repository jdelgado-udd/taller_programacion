# Entradas

Son todos aquellos datos que ingresan al programa. En Python se utiliza la función "input(str)".

```python
# Definir entrada utilizando input.
entrada = input("Ingrese algo: ")
```

La función "input(str)" te da un string, el cual puedes transformar en otros tipos de datos si es necesario.

```python
print(type(entrada))
```

# Salidas

Son absolutamente todo que sale del programa. En Python por lo general todo lo que se emita utilizando "print(str)".

```python
# Definición de variable
variable = 2

# Emitir variable, por ende ahora es una salida.
print(variable)
```
