## Pregunta 1

La implementación de este código debería de ser bastante sencilla, hay que preguntarle al usuario del programa los datos de entrada (precio unitario, costo del estampado y costo de envío) y luego revisar que cumplan las condiciones para poder ser aceptadas (que sean números, en este caso, enteros), en caso que no las cumplan, la ejecución del programa termina.

En caso de cumplir las condiciones, pasamos a la lógica principal del programa, que vendría siendo el cálculo del costo unitario (la suma de todos los datos de entrada antes solicitados), el sub-total (la multiplicación de la cantidad de poleras por el costo unitario), y el total (la multiplicación entre el sub_total, 1.19 para aplicar el IVA, y 0.92 para aplicar el 8% de descuento).

Luego de obtener todos estos datos, le mostramos al usuario cada una de las variables solicitadas en el informe mediante la función "print", en la cual incluimos strings con formato para poder insertar variables.

```python
# Este código asume que los datos ingresados están en CLP,
# por lo que no requieren decimales.
precio_unitario = input("Precio Unitario: ")
costo_estampado = input("Costo Estampado (por Un.): ")
costo_envio = input("Costo Envío (por Un.): ")
cantidad_poleras = 50 # Si se quisiera obtener una cantidad flexible, reemplazar esto por un input().

try:
    precio_unitario = int(precio_unitario)
    costo_estampado = int(costo_estampado)
    costo_envio = int(costo_envio)
except:
    print("Por favor ingrese números enteros.")
    # Esta función parece no funcionar como corresponde en Jupyter Notebook,
    # pero debería de funcionar correctamente en un archivo de Python.
    quit()

costo_unitario = (precio_unitario + costo_estampado + costo_envio) * 1.0
sub_total = cantidad_poleras * costo_unitario * 1.0
total = int(sub_total * 1.19 * 0.92) # 1.19 de IVA y 0.92 es 8% de descuento.

print("")
print(
    "Resumen del Pedido\n",
    f"Precio Unitario:              CLP {precio_unitario}\n",
    f"Costo Estampado (por Unidad): CLP {costo_estampado}\n",
    f"Costo de Envío (por Unidad):  CLP {costo_envio}\n",
    f"Costo Unitario Total:         CLP {costo_unitario}\n",
    "\n",
    f"Sub-Total:                    CLP {sub_total}\n",
    f"IVA:                          CLP {sub_total * 0.19}\n",
    f"Total (con 8% desc.):         CLP {total}\n",
    "\n",
    f"Total (por Integrante):       CLP {total / cantidad_poleras}"
)
```

## Pregunta 2

Esta implementación debería de ser más directa que la anterior, pues se solicitan 3 datos (nombre empresa como string, el cual input ya proporciona, empleados, que corresponde a un int, y promedio de almacenamiento, que corresponde a un float), debemos de revisar que 2 de estos cumplan con los requisitos para ser datos de entrada, en caso de no cumplirlos, se termina la ejecución del programa.

Pasamos a la lógica principal, donde se calculan las 3 variables solicitadas, almacenamiento necesario (una simple multiplicación entre la cantidad de empleados y el promedio de almacenamiento para obtener la cantidad total de almacenamiento), la cantidad de discos necesaria (una división entre el almacenamiento necesario y la capacidad de almacenamiento de cada disco, redondeada hacia arriba para obtener solo números enteros), y el costo mensual estimado (que corresponde a la cantidad de discos multiplicado por el costo mensual del arriendo de cada disco).

Luego se muestran los datos solicitados en pantalla utilizando la función "print".

```python
from math import ceil # Muchas gracias por permitir librerias profe.

capacidad_disco = 2000
precio_disco = 32000

nombre_empresa = input("Nombre de la empresa: ")
empleados_empresa = input("Cantidad de empleados: ")
promedio_almacenamiento = input("Promedio de almacenamiento por empleado (se permiten decimales): ")

try:
    empleados_empresa = int(empleados_empresa)
    promedio_almacenamiento = float(promedio_almacenamiento)
except:
    print("Por favor, ingrese la cantidad de empleados como un número entero y el promedio de almacenamiento como un decimal.")
    quit()

almacenamiento_necesario = empleados_empresa * promedio_almacenamiento
discos_necesarios = ceil(almacenamiento_necesario / capacidad_disco)
costo_mensual_est = discos_necesarios * precio_disco

print("")
print(
    f"Cotización empresa {nombre_empresa}\n",
    "\n",
    f"Cantidad de empleados:      {empleados_empresa}\n",
    f"Promedio de almacenamiento: {promedio_almacenamiento}GB\n",
    "\n",
    f"Almacenamiento necesario:   {almacenamiento_necesario}GB\n",
    f"Discos necesarios:          {discos_necesarios}\n",
    "\n",
    f"Costo mensual estimado:     CLP {costo_mensual_est}"
)
```
