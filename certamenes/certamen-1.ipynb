## Ejercicio 1

Entradas:
- Nombre de los 4 planetas (string)
- Índice de habitabilidad de los 4 planetas (float)

Se obtienen los valores para los 4 planetas mediante un ciclo y se va agregando el índice de habitabilidad a una variable "suma", y cada vez que un planeta tenga un índice de habitabilidad menor que 0.5, se agrega 1 a una variable "muertos". La lógica consiste en una condicional que chequea si muertos es menor que 4, en caso de serlo, clásifica al planeta mediante el valor de suma / 4, en caso de no serlo inmediatamente lo declara "Zona Muerta".

El pseudocódigo se debería de ver algo así:
```
suma = 0
muertos = 0

repetir 4 veces:
    nombre = entrada()
    habitabilidad = entrada()

    validar nombre no vacío
    validar 0 <= habitabilidad <= 1

    suma += habitabilidad

    si habitabilidad < 0.5:
        muertos += 1

si muertos == 4:
    salida("Zona Muerta")
sino:
    promedio = suma / 4

    si promedio < 0.4:
        salida("Zona Marginal")
    sino si 0.4 <= promedio <= 0.7:
        salida("Zona Moderada")
    sino:
        salida("Zona Prometedora")
```
