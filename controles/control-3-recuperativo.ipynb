## Pregunta 1
Programa de lectura de lineas
* Formato 'nombre tiempo1 tiempo2 tiempo3.
* Tiempo como float [10, 300].
* Nombre con capitalize.
* Salir con 'fin'.
* Lineas vacías (o que no cumplan) descartadas.
* Almacenar con llave "nombre" y valor "promedio" tiempos.

Reporte
* Cantidad de atletas.
* Atleta con mejor tiempo.
* Cantidad de atletas promedio menor a 60.
* Promedio general de los atletas, round 2.

```python
print("""
REGISTRO DE TIEMPOS DEPORTIVOS

Formato : nombre tiempo1 tiempo2 tiempo3
Salida  : fin
""")

atletas = {}

while True:
    entrada = input("> ").lower()

    if entrada == "fin":
        break

    if len(entrada) < 1:
        print("Por favor ingrese una entrada válida.")
        continue

    partes = entrada.split(" ")

    if len(partes) != 4:
        print("Por favor ingrese una entrada completa.")
        continue

    nombre = partes[0].capitalize()
    tiempo1 = partes[1]
    tiempo2 = partes[2]
    tiempo3 = partes[3]

    try:
        tiempo1 = float(tiempo1)
        tiempo2 = float(tiempo2)
        tiempo3 = float(tiempo3)
    except ValueError:
        print("Por favor ingrese una entrada con tiempos en formato decimal.")
        continue

    if (tiempo1 < 10 or tiempo1 > 300) or (tiempo2 < 10 or tiempo2 > 300) or (tiempo3 < 10 or tiempo3 > 300):
        print("Por favor asegúrese que los tiempos en la entrada estén dentro del rango permitido.")
        continue

    tiempo_promedio = (tiempo1 + tiempo2 + tiempo3) / 3
    atletas[nombre] = tiempo_promedio

if len(atletas) == 0:
    print("No se han registrado tiempos deportivos.")
else:
    mejor_atleta = None
    alto_rendimiento = {}
    promedio_general = sum(atletas.values()) / len(atletas)

    for atleta, tiempo in atletas.items():
        if tiempo < 60:
            alto_rendimiento[atleta] = tiempo

        if mejor_atleta is None or tiempo < atletas[mejor_atleta]:
            mejor_atleta = atleta

    print(
        "REPORTE ATLETAS",
        f" Cantidad de Atletas:                     {len(atletas)}",
        f" Atleta con mejor tiempo:                 {mejor_atleta} ({atletas[mejor_atleta]} segundos)",
        f" Cantidad de Atletas de Alto Rendimiento: {len(alto_rendimiento)}",
        f" Promedio general de los atletas:         {round(promedio_general, 2)} segundos",
        sep="\n"
    )
```

## Pregunta 2
Entrada:
* Nombres de usuario, uno a la vez.
* Salir con 'salir'.
* Vacío no se evalúa.

Proceso:
* Entre 6 y 20 caracteres.
* Comenzar con letra
* Contener al menos un dígito
* Únicamente por alpha más "._-"
* No contener 'admin' ni 'root'
* Llevar conteo válidos e inválidos
* Diccionario de fallas, largo, inicio, dígito, caracteres, prohibidos.

Salida:
* Cantidad de nombres válidos / inválidos
* Conteo de fallas por criterio

```python
print("""

VALIDADOR DE NOMBRES DE USUARIO

 (Para salir del programa, ingrese "salir")
""")

contador_validas = 0
contador_invalidas = 0

criterios = { "largo": 0, "inicio": 0, "digito": 0, "caracteres": 0, "prohibidos": 0 }

while True:
    entrada = input("Ingrese un usuario: ")

    if len(entrada) < 1:
        print("Por favor, ingrese un usuario.")
        continue

    if entrada.lower() == "salir":
        break

    largo = len(entrada)
    inicio = entrada[0].isalpha()
    digito = 0
    caracteres = 0
    prohibidos = "admin" in entrada.lower() or "root" in entrada.lower()

    for character in entrada: # True numéricamente es tratado como 1, mientras que False es tratado como 0.
        digito += character.isdigit()
        caracteres += not (character.isalnum() or character in ".-_")

    valida = True

    if largo < 6 or largo > 20:
        criterios["largo"] += 1
        valida = False

    if not inicio:
        criterios["inicio"] += 1
        valida = False

    if digito < 1:
        criterios["digito"] += 1
        valida = False

    if caracteres > 0:
        criterios["caracteres"] += 1
        valida = False

    if prohibidos:
        criterios["prohibidos"] += 1
        valida = False

    if valida:
        contador_validas += 1
    else:
        contador_invalidas += 1

contador_totales = contador_validas + contador_invalidas

if contador_totales == 0:
    print("No se ha generado un reporte puesto que no se han ingresado usuarios.")
else:
    print(
        "",
        "REPORTE DE USUARIOS",
        f" Válidas: {contador_validas} / {contador_totales}",
        f" Inválidas: {contador_invalidas} / {contador_totales}",
        f" Criterios:",
        sep="\n"
    )

    for criterio in criterios:
        print(f" - {criterio}: {criterios[criterio]} / {contador_invalidas}")
```
