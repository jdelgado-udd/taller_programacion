```python
operadores = {
    "*": lambda x, y: x * y,
    "/": lambda x, y: x / y,
    "^": lambda x, y: x ** y,
    "+": lambda x, y: x + y,
    "-": lambda x, y: x - y
}

expresion = input("Escribe una expresion simple separada por espacios (binomio): ")
expresion = expresion.split(" ")

first = 0
second = 0

try:
    first = float(expresion[0])
    second = float(expresion[2])
except:
    print("Número invalido")
    exit()

funcion = operadores[expresion[1]]

if funcion is None:
    print("Operador invalido")
    exit()

print(f"El valor de tu expresión es {funcion(first, second)}.")
```
