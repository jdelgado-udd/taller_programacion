## Pregunta 1

$3.000 por kilo envío nacional
$5.000 por kilo envío regional
$12.000 por kilo envío internacional

Mayor de 30 kilo adicional 20%
Seguro opcional adicional 15% sobre recargo

Soliticar
- Peso puede tener decimales (float)
- Destino (3 posibles valores, en string)
- Seguro (boolean)

Informe completo
- Costo base
- Recargo por peso adicional
- Seguro

```python
print("CALCULADORA DE ENVÍO")

destinos = ["regional", "nacional", "internacional"]
precios = [3000, 5000, 12000]

err = False
peso = 0

while type(peso) != float or peso <= 0:
    if err:
        print("ERROR: Peso debe de ser un número decimal mayor que 0.")
        err = False
    
    peso = input("Ingrese el peso del paquete: ")
    
    try:
        peso = float(peso)
    except:
        pass

    err = True

err = False
tipo = -1

while tipo < 0 or tipo > len(destinos) - 1:
    if err:
        print("ERROR: Debe de elegir una de las opciones.")
    
    tipo = input(f"Que tipo de envío necesita? ({destinos}) ")
    tipo = tipo.lower()

    if tipo in destinos:
        tipo = destinos.index(tipo)
    else:
        tipo = -1
    
    err = True

seguro = input("Desea agregar un seguro por un 15% adicional? (S o Cualquiera) ").lower() == "s"

base = precios[tipo]
recargo = 0

if peso > 30:
    recargo = base * 0.2

cargo_seguro = 0

if seguro:
    cargo_seguro = (base + recargo) * 0.15

final = base + recargo + cargo_seguro

print(
    f"INFORME FINAL ENVÍO [{destinos[tipo]}]",
    f" Costo base:     {base}",
    f" Peso adicional: {recargo}",
    f" Seguro:         {cargo_seguro}",
    "",
    f" TOTAL           {final}",
    sep="\n"
)
```

## Pregunta 2

Clasificar volumenes
- Pequeño [0, 0.5[
- Mediano [0.5, 3]
- Grande ]3, 10]
- Extra ]10, +]

Peso vacío: volumen * x
- Vidrio 0.8
- Plastico 0.2
- Metal 1.2

Si XL de Vidrio emitir fragil.

Solicitar:
- Volumen
- Unidad (lt o gal)
- Material

Si galones lt = gal * 3.7854

Emitir:
- Volumen en lt
- Clasificación
- Peso vacío
- Advertencia

```python
print("CLASIFICADORA DE VOLUMEN")

materiales = ["plastico", "vidrio", "metal"]
factor = [0.2, 0.8, 1.2]

err = False
volumen = 0

while type(volumen) != float or volumen <= 0:
    if err:
        print("ERROR: Volumen debe de ser un número decimal mayor que 0.")
        err = False
    
    volumen = input("Ingrese el volumen: ")
    
    try:
        volumen = float(volumen)
    except:
        pass

    err = True

err = False
unidad = ""

while unidad != "lt" and unidad != "gal":
    if err:
        print("ERROR: Debe de ingresar lt o gal.")

    unidad = input("Ingrese la unidad de volumen (lt / gal):").lower()
    err = True

err = False
material = -1

while material < 0 or material > len(materiales) - 1:
    if err:
        print("ERROR: Debe de elegir una de las opciones.")
    
    material = input(f"Ingrese el material ({materiales}) ")
    material = material.lower()

    if material in materiales:
        material = materiales.index(material)
    else:
        material = -1
    
    err = True

if unidad == "gal":
    unidad = "lt"
    volumen *= 3.7854

clasificacion = ""

if volumen < 0.5:
    clasificacion = "pequeño"

if volumen >= 0.5:
    clasificacion = "mediano"

if volumen > 3:
    clasificacion = "grande"

if volumen > 10:
    clasificacion = "extra grande"

pesovacio = volumen * factor[material]
advertencia = ""

if materiales[material] == "vidrio" and clasificacion == "extra grande":
    advertencia = "ADVERTENCIA: El envase es muy frágil, por ser de vidrio y tamaño XL."

print(
    "INFORME",
    f" Volumen:       {volumen}lt",
    f" Peso Vacío:    {pesovacio}",
    f" Clasificación: {clasificacion}",
    f" {advertencia}",
    sep="\n"
)
```
